```python
import numpy
import cv2
from matplotlib import pyplot as plt
```

```python
image1 = cv2.imread("cloudnerd-demo.tiff")
image2 = cv2.imread("shazoom.tiff")
image3 = cv2.imread("chatbot.png")

w,h,c = image3.shape
w = w
h = h
```

```python
image1 = cv2.resize(image1, (h,w), interpolation=cv2.INTER_CUBIC)
image2 = cv2.resize(image2, (h,w), interpolation=cv2.INTER_CUBIC)
image3 = cv2.resize(image3, (h,w), interpolation=cv2.INTER_CUBIC)
```

```python
print(image1.shape)
print(image2.shape)
print(image3.shape)
```

```python
cv2.imwrite("shazoom.png", image2)
cv2.imwrite("cloudnerd-demo.png", image1)
cv2.imwrite("chatbot.png", image3)
```

```python
plt.subplot(1,3,1)
plt.imshow(image1)
plt.subplot(1,3,2)
plt.imshow(image2)
plt.subplot(1,3,3)
plt.imshow(image3)
```
